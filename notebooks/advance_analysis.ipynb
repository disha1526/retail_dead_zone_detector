```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../data/cleaned_superstore.csv')

# Average profit for each discount bracket
discount_impact = df.groupby('Discount_Bracket').agg(
    Avg_Profit  = ('Profit', 'mean'),
    Avg_Sales   = ('Sales', 'mean'),
    Orders      = ('Sales', 'count'),
    Loss_Rate   = ('Is_Loss', 'mean')
).reset_index()

discount_impact['Loss_Rate'] = round(discount_impact['Loss_Rate'] * 100, 1)

print("💡 KEY INSIGHT — How Discounts Kill Profit:")
print(discount_impact)

# This will show: higher discount = more losses!
```

```python
# Region-wise dead zone analysis
region_summary = df.groupby(['Region', 'Category']).agg(
    Sales   = ('Sales', 'sum'),
    Profit  = ('Profit', 'sum'),
    Margin  = ('Profit_Margin', 'mean')
).reset_index()

print("🗺️ Region + Category Performance:")
display(region_summary.sort_values('Profit'))
```

```python
# Which customer segment is most profitable?
segment = df.groupby('Segment').agg(
    Sales     = ('Sales', 'sum'),
    Profit    = ('Profit', 'sum'),
    Orders    = ('Sales', 'count'),
    Margin    = ('Profit_Margin', 'mean')
).reset_index()

segment['Profit_per_Order'] = round(segment['Profit'] / segment['Orders'], 2)
print("👥 Segment Analysis:")
display(segment)
```

```python
# Does fast shipping hurt profits?
ship = df.groupby('Ship Mode').agg(
    Avg_Profit = ('Profit', 'mean'),
    Orders     = ('Sales', 'count'),
    Avg_Sales  = ('Sales', 'mean')
).reset_index().sort_values('Avg_Profit', ascending=False)

print("🚚 Ship Mode vs Profit:")
display(ship)
```
